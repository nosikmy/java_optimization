```python
import matplotlib.pyplot as plt

with open("memory_usage.txt", "r") as file:
    lines = file.readlines()

max_memory = []
allocated_memory = []
used_memory = []
free_memory = []

for line in lines:
    data = line.split()
    max_memory.append(int(data[0]))
    allocated_memory.append(int(data[1]))
    used_memory.append(int(data[2]))
    free_memory.append(int(data[3]))

plt.figure(figsize=(10, 6))
plt.plot(max_memory, label='Max Memory')
plt.plot(allocated_memory, label='Allocated Memory')
plt.plot(used_memory, label='Used Memory')
plt.plot(free_memory, label='Free Memory')
plt.xlabel('Iterations')
plt.ylabel('Memory (in bytes)')
plt.title('Memory Usage Over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
